```python
import matplotlib .pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = [1,2,3,4,5,6,7]
y = [34,23,34,12,25,16,38]

plt.plot(x,y, color = "red" , alpha = 1 , marker = "+")
plt.show()
```

```python
days = [1,2,3,4,5,6,7]
max_temperature = [45,34,43,37,36,41,40]
min_temperature = [20,21,25,30,29,18,13]
avg_temperature = [32,27,34,33,32,29,26]

plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("7 Days Weather Graph")

plt.plot(days,max_temperature , label = "Max" , marker = ".")
plt.plot(days,min_temperature , label = "Min" , marker = ".")
plt.plot(days,avg_temperature , label = "Avg" , marker = ".")

plt.legend("10")

plt.grid()
```

## Bar Chart

```python
company = ["Bashundhara" , "Beximco" , "Walton" , "Beximco" , "Square"]
revenu = [60,50,70,55,70]
profit = [20,40,50,10,30]

xposition = np.arange(len(company))
xposition

plt.bar(xposition-0.2,revenu , label = "reveneu" , width = 0.4)
plt.bar(xposition+0.2,profit , label = "profit" , width = 0.3)

plt.xticks(xposition,company)

plt.xlabel("Company Name")
plt.ylabel("Revenu(dollar)")
plt.title("company Profit")
plt.legend()
plt.grid()
```

## Histogram 

```python
blood_suger_men = [120,130,111,114,130,120,90,70,80,84,92]
blood_suger_women = [70,92,111,84,115,122,95,110,60,88,82]

plt.hist([blood_suger_men,blood_suger_women] , bins=[60,80,120,150] , rwidth = 0.9 , label=["men","women"] , color=["r","g"])

plt.xlabel("Blood Suger level")
plt.ylabel("Total number of patients")
plt.title("Blood Suger analysis")
plt.legend()
plt.grid()
```

## Pie Chart
